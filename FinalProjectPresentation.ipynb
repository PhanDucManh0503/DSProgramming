👉 Mục tiêu:

Tìm giá trị nhỏ nhất (min), lớn nhất (max) và độ lệch chuẩn (std) để xác định sự phân phối của dữ liệu.
Phát hiện outliers (giá trị bất thường) nếu min/max quá chênh lệch với giá trị trung bình (mean).

📊 Giải thích:
min và max cho biết phạm vi của dữ liệu.
mean phản ánh giá trị trung bình, nếu max quá lớn có thể là dấu hiệu của outliers.

```python
import pandas as pd
import numpy as np

# Đọc dữ liệu
data = pd.read_csv('wdbc.data', header=None)

# Đặt tên cột cho dễ hiểu
data.columns = ['ID', 'Diagnosis'] + [f'Feature_{i}' for i in range(1, 31)]

# Kiểm tra kiểu dữ liệu của các cột
print("Kiểu dữ liệu của từng cột:\n")
print(data.dtypes)
```

```python
# Mã hóa cột Diagnosis (M: 1, B: 0)
data['Diagnosis'] = data['Diagnosis'].map({'M': 1, 'B': 0})

# Loại bỏ cột ID (không phải đặc trưng số học)
data.drop(columns=['ID'], inplace=True)
```

```python
# Lọc chỉ các cột có kiểu số (float64, int64)
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Kiểm tra lại kiểu dữ liệu sau khi lọc
print("\nDữ liệu sau khi lọc chỉ các cột số:\n")
print(numerical_data.dtypes)
```

```python
# Tính toán phân vị (IQR)
Q1 = numerical_data.quantile(0.25)  # Phân vị thứ 25%
Q3 = numerical_data.quantile(0.75)  # Phân vị thứ 75%
IQR = Q3 - Q1  # Khoảng giá trị giữa Q3 và Q1

# Xác định outliers (giá trị nằm ngoài [Q1 - 1.5*IQR, Q3 + 1.5*IQR])
outliers = (numerical_data < (Q1 - 1.5 * IQR)) | (numerical_data > (Q3 + 1.5 * IQR))

# Đếm số lượng outliers trong mỗi cột
print("\nSố lượng outliers trong mỗi cột:\n")
print(outliers.sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Vẽ boxplot để trực quan hóa outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=numerical_data)
plt.xticks(rotation=90)
plt.title("Boxplot - Phát hiện outliers trong từng cột")
plt.show()
```

Feature_4 (Mean Area) và Feature_24 (Worst Area) là hai đặc trưng quan trọng với nhiều outliers.
Các đặc trưng này có thể là yếu tố quyết định giúp phân biệt khối u lành tính và ác tính.
Outliers trong y tế có ý nghĩa đặc biệt, không nên loại bỏ hoàn toàn mà cần cân nhắc kỹ lưỡng.

##### 2.6 Phân tích chi tiết: Phân bố dữ liệu cột phân loại và kiểm tra missing values, các giá trị khác nhau và giá trị bất thường (abnormal)

**Kiểm tra giá trị bị thiếu (missing values) trong cột phân loại.**

```python
# Kiểm tra giá trị bị thiếu trong cột phân loại
missing_values = data.isnull().sum()
missing_percentage = (missing_values / len(data)) * 100

# Kiểm tra tất cả cột phân loại (kể cả int và object)
categorical_columns = data.select_dtypes(include=['object', 'int']).columns
print("Tỷ lệ phần trăm giá trị bị thiếu trong các cột phân loại:\n")
print(missing_percentage[categorical_columns])
```

Không có giá trị thiếu, kết quả trả về là Series rỗng.

**Đếm số lượng các giá trị khác nhau**

```python
# Đếm số lượng giá trị khác nhau trong cột phân loại
for col in categorical_columns:
    print(f"\nCột {col} có {data[col].nunique()} giá trị khác nhau.")
    print(f"Ví dụ một vài giá trị:\n{data[col].unique()[:5]}")
```

**Phát hiện dữ liệu bất thường**

```python
for col in categorical_columns:
    print(f"\nPhân bố giá trị trong cột {col}:\n")
    print(data[col].value_counts())
```
