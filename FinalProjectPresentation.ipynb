##### 2.5 Ph√¢n t√≠ch chi ti·∫øt: Ph√¢n b·ªë d·ªØ li·ªáu c·ªôt s·ªë v√† ki·ªÉm tra missing values, gi√° tr·ªã b·∫•t th∆∞·ªùng (abnormal)

- **Missing values (Gi√° tr·ªã thi·∫øu)** c√≥ th·ªÉ ·∫£nh h∆∞·ªüng ƒë·∫øn ƒë·ªô ch√≠nh x√°c c·ªßa m√¥ h√¨nh h·ªçc m√°y. Vi·ªác x√°c ƒë·ªãnh v√† x·ª≠ l√Ω missing values l√† m·ªôt ph·∫ßn quan tr·ªçng trong ti·ªÅn x·ª≠ l√Ω d·ªØ li·ªáu.  
- **Gi√° tr·ªã b·∫•t th∆∞·ªùng (Abnormal values)** c√≥ th·ªÉ l√†m l·ªách ph√¢n ph·ªëi d·ªØ li·ªáu, d·∫´n ƒë·∫øn m√¥ h√¨nh h·ªçc sai ho·∫∑c kh√¥ng hi·ªáu qu·∫£. C√°c gi√° tr·ªã qu√° l·ªõn ho·∫∑c qu√° nh·ªè so v·ªõi m·∫∑t b·∫±ng chung c·∫ßn ƒë∆∞·ª£c ph√°t hi·ªán v√† x·ª≠ l√Ω.

**C√°c b∆∞·ªõc ph√¢n t√≠ch v√† x·ª≠ l√Ω d·ªØ li·ªáu:**

üëâ M·ª•c ti√™u:

T√¨m gi√° tr·ªã nh·ªè nh·∫•t (min), l·ªõn nh·∫•t (max) v√† ƒë·ªô l·ªách chu·∫©n (std) ƒë·ªÉ x√°c ƒë·ªãnh s·ª± ph√¢n ph·ªëi c·ªßa d·ªØ li·ªáu.
Ph√°t hi·ªán outliers (gi√° tr·ªã b·∫•t th∆∞·ªùng) n·∫øu min/max qu√° ch√™nh l·ªách v·ªõi gi√° tr·ªã trung b√¨nh (mean).

üìä Gi·∫£i th√≠ch:
min v√† max cho bi·∫øt ph·∫°m vi c·ªßa d·ªØ li·ªáu.
mean ph·∫£n √°nh gi√° tr·ªã trung b√¨nh, n·∫øu max qu√° l·ªõn c√≥ th·ªÉ l√† d·∫•u hi·ªáu c·ªßa outliers.

```python
import pandas as pd
import numpy as np

# ƒê·ªçc d·ªØ li·ªáu
data = pd.read_csv('wdbc.data', header=None)

# ƒê·∫∑t t√™n c·ªôt cho d·ªÖ hi·ªÉu
data.columns = ['ID', 'Diagnosis'] + [f'Feature_{i}' for i in range(1, 31)]

# Ki·ªÉm tra ki·ªÉu d·ªØ li·ªáu c·ªßa c√°c c·ªôt
print("Ki·ªÉu d·ªØ li·ªáu c·ªßa t·ª´ng c·ªôt:\n")
print(data.dtypes)
```

```python
# M√£ h√≥a c·ªôt Diagnosis (M: 1, B: 0)
data['Diagnosis'] = data['Diagnosis'].map({'M': 1, 'B': 0})

# Lo·∫°i b·ªè c·ªôt ID (kh√¥ng ph·∫£i ƒë·∫∑c tr∆∞ng s·ªë h·ªçc)
data.drop(columns=['ID'], inplace=True)
```

```python
# L·ªçc ch·ªâ c√°c c·ªôt c√≥ ki·ªÉu s·ªë (float64, int64)
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Ki·ªÉm tra l·∫°i ki·ªÉu d·ªØ li·ªáu sau khi l·ªçc
print("\nD·ªØ li·ªáu sau khi l·ªçc ch·ªâ c√°c c·ªôt s·ªë:\n")
print(numerical_data.dtypes)
```

```python
# T√≠nh to√°n ph√¢n v·ªã (IQR)
Q1 = numerical_data.quantile(0.25)  # Ph√¢n v·ªã th·ª© 25%
Q3 = numerical_data.quantile(0.75)  # Ph√¢n v·ªã th·ª© 75%
IQR = Q3 - Q1  # Kho·∫£ng gi√° tr·ªã gi·ªØa Q3 v√† Q1

# X√°c ƒë·ªãnh outliers (gi√° tr·ªã n·∫±m ngo√†i [Q1 - 1.5*IQR, Q3 + 1.5*IQR])
outliers = (numerical_data < (Q1 - 1.5 * IQR)) | (numerical_data > (Q3 + 1.5 * IQR))

# ƒê·∫øm s·ªë l∆∞·ª£ng outliers trong m·ªói c·ªôt
print("\nS·ªë l∆∞·ª£ng outliers trong m·ªói c·ªôt:\n")
print(outliers.sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# V·∫Ω boxplot ƒë·ªÉ tr·ª±c quan h√≥a outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=numerical_data)
plt.xticks(rotation=90)
plt.title("Boxplot - Ph√°t hi·ªán outliers trong t·ª´ng c·ªôt")
plt.show()
```

C√°c ƒë·∫∑c tr∆∞ng n√†y c√≥ th·ªÉ l√† y·∫øu t·ªë quy·∫øt ƒë·ªãnh gi√∫p ph√¢n bi·ªát kh·ªëi u l√†nh t√≠nh v√† √°c t√≠nh.
Outliers trong y t·∫ø c√≥ √Ω nghƒ©a ƒë·∫∑c bi·ªát, kh√¥ng n√™n lo·∫°i b·ªè ho√†n to√†n m√† c·∫ßn c√¢n nh·∫Øc k·ªπ l∆∞·ª°ng.

##### 2.6 Ph√¢n t√≠ch chi ti·∫øt: Ph√¢n b·ªë d·ªØ li·ªáu c·ªôt ph√¢n lo·∫°i v√† ki·ªÉm tra missing values, c√°c gi√° tr·ªã kh√°c nhau v√† gi√° tr·ªã b·∫•t th∆∞·ªùng (abnormal)

**Ki·ªÉm tra gi√° tr·ªã b·ªã thi·∫øu (missing values) trong c·ªôt ph√¢n lo·∫°i.**

```python
# Ki·ªÉm tra gi√° tr·ªã b·ªã thi·∫øu trong c·ªôt ph√¢n lo·∫°i
missing_values = data.isnull().sum()
missing_percentage = (missing_values / len(data)) * 100

# Ki·ªÉm tra t·∫•t c·∫£ c·ªôt ph√¢n lo·∫°i (k·ªÉ c·∫£ int v√† object)
categorical_columns = data.select_dtypes(include=['object', 'int']).columns
print("T·ª∑ l·ªá ph·∫ßn trƒÉm gi√° tr·ªã b·ªã thi·∫øu trong c√°c c·ªôt ph√¢n lo·∫°i:\n")
print(missing_percentage[categorical_columns])
```

Kh√¥ng c√≥ gi√° tr·ªã thi·∫øu, k·∫øt qu·∫£ tr·∫£ v·ªÅ l√† Series r·ªóng.

**ƒê·∫øm s·ªë l∆∞·ª£ng c√°c gi√° tr·ªã kh√°c nhau**

```python
# ƒê·∫øm s·ªë l∆∞·ª£ng gi√° tr·ªã kh√°c nhau trong c·ªôt ph√¢n lo·∫°i
for col in categorical_columns:
    print(f"\nC·ªôt {col} c√≥ {data[col].nunique()} gi√° tr·ªã kh√°c nhau.")
    print(f"V√≠ d·ª• m·ªôt v√†i gi√° tr·ªã:\n{data[col].unique()[:5]}")
```

**Ph√°t hi·ªán d·ªØ li·ªáu b·∫•t th∆∞·ªùng**

```python
for col in categorical_columns:
    print(f"\nPh√¢n b·ªë gi√° tr·ªã trong c·ªôt {col}:\n")
    print(data[col].value_counts())
```
